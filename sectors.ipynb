```python
import yfinance as yf
import os
import pandas as pd
import hvplot.pandas
import numpy as np

%matplotlib inline
```

```python
# List of S&P 500 sector ETFs
etfs = ["XLK","XLY","XLC","XLI","XLU","XLRE","XLB","XLP", "XLF", "XLV", "XLE","DX-Y.NYB"]
```

```python
# Download the ETF data from Yahoo Finance
data = yf.download(etfs, start="2008-01-01", end="2022-12-31", interval="1d")["Close"]
```

```python
sectors_data= pd.DataFrame(data).drop(columns=["XLC","XLRE"])
```

```python
sectors_data.head()
```

```python
columns=["US Dollar","Materials" ,"Energy", "Financials", "Industrials","Info Tech","Consumer Staples","Utilities","Healthcare","Consumer Desc"]
```

```python
sectors_data.columns=columns
```

```python
sectors_data.head()
```

```python
sectors_data.hvplot.line(
    xlabel="Year",
    ylabel="Closing Price"
)
```

```python
returns=sectors_data.pct_change().dropna()
```

```python
returns.head()
```

```python
#Determine Fibonacci Ratios 

def fibonacci_ratios(sectors_data):
    ret = []
    for i in range(len(sectors_data) - 1):
        ret.append((sectors_data[i + 1] - sectors_data[i]) / sectors_data[i])
    ret = np.array(ret)
    ret_mean = np.mean(ret)
    ret_std = np.std(ret)
    ratio = []
    for r in ret:
        ratio.append((r - ret_mean) / ret_std)
    return ratio
```

```python
# calculate fibonacci ratios for each ETF
fib_ratios = {}
for col in sectors_data.columns:
    fib_ratios[col] = fibonacci_ratios(sectors_data=[col])

print(fib_ratios)
```

```python
#2ND METHOD


# Calculate Fibonacci retracements for each ETF based on the most recent high and low
fib_ratios = [0, 0.236, 0.382, 0.5, 0.618, 1]
fib_levels = {}
for etf in etfs:
    high, low = returns[etf].max(), returns[etf].min()
    fib_levels[etf] = [high - x * (high - low) for x in fib_ratios]

# Compare each ETF's daily returns to its Fibonacci levels
for etf in etfs:
    returns[f'{etf}_fib'] = returns[etf].apply(lambda x: 'above' if x > fib_levels[etf][-1] else
                                                'below' if x < fib_levels[etf][0] else
                                                'between' if any([x > level for level in fib_levels[etf][:-1]]) else
                                                'on')

# Print the results
print(returns)

```

```python
#3RD METHOD

# Calculate Fibonacci ratios (23.6%, 38.2%, 50%, 61.8%, 100%)
fib_ratios = [0.236, 0.382, 0.5, 0.618, 1.0, 1.618]
fib_levels = []
for ratio in fib_ratios:
    fib_level = np.percentile(returns, ratio * 100)
    fib_levels.append(fib_level)

    
# Print results
print("Fibonacci Ratios:")
for ratio, level in zip(fib_ratios, fib_levels):
    print(f"{ratio * 100:.1f}%: {level:.4f}")
```

```python
# CONVERSION OF FIBONACCI RATIOS TO PRICES TO STOCHASTIC

ef fibonacci_ratios_to_prices(high, low, retracements):
    prices = []
    for retracement in retracements:
        price = low + (high - low) * retracement
        prices.append(price)
    return prices
```

```python
def prices_to_stochastic(prices, close):
    stochastic = []
    for price in prices:
        stochastic.append((close - price) / (close - low))
    return stochastic
```

```python
prices = fibonacci_ratios_to_prices(high, low, retracements)
stochastic = prices_to_stochastic(prices, close)
```

```python
print("Prices:", prices)
print("Stochastic:", stochastic)
```
